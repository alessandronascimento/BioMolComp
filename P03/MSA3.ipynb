Uma ferramenta que permite fazer isto é o [Clustal Omega](https://www.ebi.ac.uk/Tools/msa/clustalo/). 

Nesta atividade usaremos o Clustal Omega *online* para alinhar as sequências de 48 membros da família dos receptores nucleares.



1.   Abra o [Clustal Omega](https://www.ebi.ac.uk/Tools/msa/clustalo/) em uma nova aba do navegador.
2.   Copie as sequências abaixo e cole no formulário do Clustal Omega, no campo "*sequences in any supported format:*" (as sequências estão em formato FASTA).
3. Clique em "*Submit*"
