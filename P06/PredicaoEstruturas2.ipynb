## Modelagem Por Homologia ##

O programa que vamos usar chama-se [Modeller](https://salilab.org/modeller/). Foi desenvolvido por Andrej Sali e Tom Blundell nos anos 1990. O programa ainda é ativamente desenvolvido e mantido e os desenvolvedores ainda mantém um servidor que permite a execução da modelagem de forma *online* e relativamente rápida. 

O programa e o servidor são mantidos pela Universidade da Califórnia em San Francisco (UCSF). A mesma instituição também desenvolve um programa de visualização de estruturas chamado [UCSF Chimera](https://www.cgl.ucsf.edu/chimera/). O Chimera permite visualizar e manipular modelos estruturais. Tem ainda uma ferramenta que permite executar a modelagem com o Modeller de forma simples. Nós usaremos as duas ferramentas nesta etapa da prática. 


## Construção do Modelo (Sistema Linux)

1.   Abra o programa *UCSF Chimera* ;
2.   Vá em **Tools** -> **Sequence** -> **Blast Protein** ; 
3.   Na janela que se abrirá, selecione a aba **Plain Text**. Copie e cole a sequência da nossa proteína alvo na caixa de sequência e clique em **OK**.
4.   Você provavelmente vai observar um resultado similar a este mostrado abaixo como resultado da sua busca.

O BLAST nos mostra que a proteína depositada no PDB com código 5MLZ (cadeia A) é o homólogo mais próximo da nossa sequência alvo, com um e-value de $1.9 \times 10^{-9}$.

5. Selecione as três primeiras sequências (5MLZ, 5MM0 e 5MM1) e clique em **Show in MAV**, para visualizar o alinhamento múltiplo destas sequências.

Neste ponto, vemos as sequências que foram alinhadas à nossa sequência alvo. Como já havíamos percebido na primeira parte da prática, o alinhamento não cobre toda a extensão da nossa sequência. Além disso, as três sequências homólogas à nossa são idênticas entre si. Na prática, temos apenas uma sequência de estrutura conhecida que é homóloga à nossa sequência alvo (embora tenha baixa identidade) e com uma cobertura limitada. Este é um desafio bem grande para a modelagem por homologia!

6. Na janela do MAV, vá em **Structure** -> **Modeller (homology)**.  
7. Na janela que se abrirá temos a opção de selecionar o *template* (molde) que usaremos para a modelagem por homologia. Também podemos definir o número de modelos que queremos gerar. Finalmente, podemos tambem definir algumas opções mais avançadas, se assim desejarmos.
8. Selecione os três *templates*. Veja que o Chimera já entendeu que a sequência que será modelada é a sequência *query*, que usamos como busca no BLAST.
9. Entre com a *chave de licença* e clique em **OK**.

Screen Shot 2022-05-18 at 11.47.23.png

A geração dos modelos deve ser finalizada em poucos minutos. Em seguida, os modelos serão mostrados na janela principal do Chimera. Observe dois achados importantes aqui.

* Uma janela será aberta mostrando os resultados do modeller. Cinco modelos foram gerados. Estes modelos foram avaliados usando um potencial estatístico chamdo DOPE (Discrete Optimized Protein Energy). Como o potencial DOPE é dependente do tamnho da proteína, este potencial é normalizado na forma de Z-scores (zDOPE). Quanto menor o valor de zDOPE, melhor o modelo. Um zDOPE de -1.5 representa um modelo que é 1.5 desvio-padrão melhor que média.  Veja que nossos modelos não têm valores muito interessantes de zDOPE. A mesma coisa pode ser vista com avaliação chamada GA341. Usualmente, um valor de GA341 > 0.7 indica um modelo razoável.
