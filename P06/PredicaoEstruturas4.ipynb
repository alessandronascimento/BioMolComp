# Análise de Estruturas/Modelos #

Há muitas formas de analisar um modelo ou uma estrutura gerada. Uma forma interessante é comparar as coordenadas de um modelo com todos os 180.000+ modelos existentes no [PDB](https://www.rcsb.org/). 

O servidor [DALI](http://ekhidna2.biocenter.helsinki.fi/dali/) permite fazer esta comparação. O servidor reporta a qualidade da superposição das coordenadas atômicas do nosso modelo com outros modelos existentes no [PDB](https://www.rcsb.org/) na forma e um Z-score. Quanto maior o Z-score, tanto melhor a superposição das coordenadas, i.e., tanto mais estruturalmente similar as estruturas são.

Neste servidor, clique em **PDB Search**.

Screen Shot 2022-05-18 at 12.27.43.png

1.   Faça o upload do modelo que obtivemos com o AlphaFold. 
2.   Dê um nome para a sua busca (por exemplo, MeuNome_BMC)
3.   Coloque seu email no formulário. Você receberá um email quando a busca estiver completa.
4.  Clique em **Submit**.

A comparação pode demorar algumas horas, dependendo de quantos jobs o servidor tem na fila. Eu fiz uma busca nesta semana. Os resultados da minha busca podem ser visualizado [neste link](http://ekhidna2.biocenter.helsinki.fi/barcosel/tmp//b20836305617d88326cd9786830600bdcf488787ce16f6d22b/). Meu job tinha o título de "*avgepaI*".


Screen Shot 2022-05-18 at 12.31.03.png

5. Clique em **Matches against PDB90**. Este link vai mostrar os resultados da busca do nosso modelo contra o PDB filtrado para excluir sequências que são mais do que 90% idênticas. 
6. Nesta página veremos resultados como os mostrados abaixo. Para cada *match* identificado pelo DALI, temos um score Z (Z-score), o valor de RMSD (*root mean square deviation*) entre o nosso modelo e a estrutura existente no PDB. Se as estruturas fossem idênticas, o RMSD seria 0.0. Neste caso, quanto menor o RMSD, mais similares são as estruturas protéicas.
7. Veja que as estruturas com Z-score maior que 15 têm algum tipo de conservação de função. Todas estão envolvidas na síntese/adição de açúcares a compostos hidrofóbicos aceptores. Estas comparações são úteis na inferência da função da estrutura da proteína, caso ela não seja conhecida *a priori*.

# Detecção de Bolsões (Pockets) #

Eventualmente, pode ser muito útil identificar a presença de bolsões na estrutura/modelo de uma proteína. A presença destes bolsões pode indicar, por exemplo, que a proteína é efetivamente uma enzima, com um sítio de interação com um substrato, por exemplo. Outra possibilidade é que a proteína seja um receptor farmacológico. A interação com ligantes/substratos pode até mesmo ser usada como métrica para a descoberta de funções de proteínas! [Este artigo](https://www.nature.com/articles/nature05981) traz um exemplo muito legal sobre isso. 

Nesta etapa, vamos analisar o nosso modelo quanto à presença de bolsões.



1.   Usando um navegador de internet, vá até o servidor [Protein Plus](https://proteins.plus/).
2.   Na opção de "**Upload Protein (PDB format)**:", submeta o arquivo obtido como modelo com o AlphaFold e clique em clique em **Go!**.
3. A janela que se abrirá permite a manipulação/visualização do modelo online (via browser). Na parte direita da janela (veja exemplo abaixo) contem ferramentas de análise do nosso modelo. Clique na aba de **DogSiteScorer** e depois no botão **DogSiteScorer**. 
4. Mantenha as configurações padrão e clique em **Calculate**.



Screen Shot 2022-05-18 at 12.47.25.png

O cálculo deve durar alguns segundos, e o resultado será mostrado em seguida (veja um exemplo abaixo).



*   O programa identificou 5 bolsões, com volumes diferentes. O maior destes bolsões tem um volume de ~1450 Å$^3$. O programa atribui um "Drug Score" elevado para este bolsão (0.8), sugerindo que este pode ser um bolsão explorado para o desenho de fármacos!
*   Usando o botão com desenho de um olho, você pode visualizar os demais bolsões.
*   Curiosamente, esta enzima está efetivamente associada à adição de açúcares à aceptores hidrofóbicos. A enzima de maior similaridade com estrutura conhecida (PDB 5MLZ) possui o sítio ativo na posição identificada pelo DogSiteScorer. Efetivamente, a combinação AlphaFold+DogSiteScorer parece ter identificado um bom modelo e uma identificação correta da região do sítio ativo da enzima.
