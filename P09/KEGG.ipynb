# Análise de Redes Metabólicas com o KEGG #

 O KEGG é um banco de dados desenhado para se entender funções e utilidades de alto nível do sistema biológico, como a célula, o organismo e o ecossistema, a partir de informações de nível molecular, especialmente conjuntos de dados moleculares de grande escala gerados pelo sequenciamento do genoma e outros dados de alto rendimento tecnologias experimentais.

 Dentre as muitas funções do KEGG, é possível mapear redes metabólicas ou *pathways*.

 


 

## Redes de Doenças Humanas ##

 Dentre as muitas redes presentes no KEGG, existe uma categoria de doenças humanas. Esta categoria lista genes/proteínas envolvidas no processo de estabelecimento de doenças importantes que vão desde a asma até doenças infecciosas.

## Vias de Síntese de Antibióticos em Bactérias ##

 Muitos dos antibióticos que hoje empregamos foram inicialmente identificados em microrganismos, onde são usados como forma de defesa contra outros organismos que possam colonizar o mesmo ambiente. 

 Um exemplo neste contexto é o ácido clavulânico (figura abaixo). O ácido clavulânico não é um antibiótico formalmente, mas por ser um inibidor da enzima beta-lactamase, ele impede que outros antibióticos (beta lactâmicos, como a penicilina) sejam inativados. Desta forma, ele é um adjuvante na terapia com antibióticos.
