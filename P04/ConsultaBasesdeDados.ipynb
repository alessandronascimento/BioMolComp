# Consulta em Bases de Dados #

Nesta seção faremos a consulta na base de dados do NCBI através do portal de consulta Entrez *online*. Na segunda parte da prática faremos a mesma consulta através do Biopython.

## Coronavírus SARS-Cov2 ##

Como estamos no meio de uma pandemia da COVID-19, causada pelo vírus *SARS-Cov2*, vamos empregar o portal do NCBI para levantar todas as informações biológicas disponíveis sobre este vírus.

### 1. Buscando o genoma do SARS-Cov2.



1.   Vá até o portal do [NCBI](https://www.ncbi.nlm.nih.gov/) e busque pelo termo *SARS-Cov2*. 
2.   A janela que vai se abrir (exemplo abaixo) mostra que há muitas informações em várias das bases de dados do NCBI consistentes com o nosso termo de busca. Vamos selecionar a base de **Genoma**.

Screen Shot 2022-04-30 at 17.23.24.png

Na janela relativa aos genomas, temos informações referentes ao genoma do vírus SARS-Cov2 (veja o exemplo abaixo).

Dentre as informações listadas pela base de dados, já vemos que o sequenciamento foi gerado por 92 montagens de 41 *reads*. Também vemos que o conteúdo mediano de GC é de 38%. A tabela abaixo mostra o tamanho do genoma (29.9 Kb) e temos também o link para a sequência na base de dados de nucleotídeos. O código deste genoma na base de nucleotídeos é o [NC_045512.2](https://www.ncbi.nlm.nih.gov/nuccore/NC_045512.2).

Veja que esta entrada pertence ao conjunto de sequências de referência (RefSeq). Isto é, o genoma do vírus isolado em Wuhan (*Wuhan-Hu-1*) é a sequência de referência para o genoma do SARS-Cov2.

Uma forma mais simples de se realizar uma análise inicial da sequência é usando a ferramenta gráfica do próprio NCBI. Para isto, clique em *Graphics* no início da página. Você provavelmente vai encontrar uma página similar a esta que estou monstrando na figura abaixo.

Aqui vemos que o genoma do vírus codifica uma sequência longa (uma única ORF), com quase 22 Kb. Esta poliproteína é depois fragmentada em diversas proteínas mostradas no gráfico.

Em seguida (22-24Kb), temos o gene da proteína *Spike*. O genome é completo com outras proteínas estruturais codificadas na região mostrada entre 24-29.9 Kb.

### 2. Análise de uma das proteínas virais

Uma das proteínas codificadas pela primeira ORF (poliproteína) é a proteína YP_009725311.1, anotada como uma "*2'-O-ribose methyltransferase*"



1.   Identifique esta proteína na ferramenta de visualização gráfica do genoma e clique em *Graphical View YP_009725311.1*.
2.   Na aba da análise gráfica da proteína, alterne a visão para um report em *GenPept*. Ao fazer isto, você deve estar enxergando uma tela similar à tela mostrada na figura abaixo.
3. Vamos fazer algumas análises com esta proteína! Primeiramente, clique em "*Identify Conserved Domains*", no lado direito do seu navegador. 
4. **Q1: Você consegue obter alguma identificação positiva para domínios presentes na sequência desta proteína? Se sim, qual a sua função predita?**
5. Voltando à tela da sequência da proteína YP_009725311.1, clique em "*Run BLAST*" para buscarmos sequências homólogas a esta presentes em todas as bases de dados do NCBI.
6. Na janela do programa BLAST que será aberta, selecione a base de dados "*RefSeq*", de proteínas anotadas.
