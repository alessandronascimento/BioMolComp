# Análise Filogenética de Sequências #

Nesta atividade iremos realizar a análise filogenética do domínio transcriptase reversa de retrotransposons do tipo non-LTR. Este tipo de elemento transponível pode ser classificado em diversas famílias baseado na evolução do seu domínio da transcriptase reversa. O objetivo desta atividade é realizar a classificação de retrotransposons problemas em relação a estas famílias.



# Atividade 1 #

## Atividade a ser realizada no sistema *Linux* ##


1.   Realize o download do arquivo [filogenia1.fasta](https://raw.githubusercontent.com/alessandronascimento/BioMolComp/main/P05/filogenia1.fasta) e salve em uma pasta no seu disco. O arquivo contém um conjunto de sequências que usaremos na nossa análise. A primeira etapa será gerar um alinhamento múltiplo destas sequências.
2.   Abra o programa **MEGA11**. Clique em **Align** e selecione** Edit/Build Alignment**. Na janela que será aberta, selecione **Retrieve sequences from a file**. Selecione o arquivo *filogenia1.fasta*, que você baixou no item 1. 
3.   Na janela do *alignment explorer*, podemos ver que temos 31 sequências de domínios transcriptase reversa de retrotransposons do tipo non-LTR. Veja que as sequências têm tamanhos diferentes e ainda precisam ser alinhadas. 
4. Clique em **Alignment** e em **Align by ClustalW**. Vamos usar um peso para a abertura de gaps de 5.0 e um peso para a extensão de um gap de 0.1.
5. Compare o alinhamento obtido com o alinhamento mostrado na figura abaixo. Você consegue identificar os mesmos motivos conservados?

6. Se você obteve um alinhamento com os motivos conservados, salve o seu alinhamento: vá em **Data**, e depois em **Export Alignment** -> **MEGA Format**. Salve como *filogenia1.meg*. Você pode fechar o *alignment explorer* e retornar à janela principal do *MEGA11*.
7. Na janela principal do programa MEGA11, vá em **Phylogeny** e depois clique em **Construct/Test Neighbor-joining tree**.
8. O *MEGA11* vai perguntar qual o alinhamento que será usado na construção da árvore. Selecione o arquivo com o alinhamento que geramos: *filogenia1.meg*.
9. No campo de **Test of Phylogeny**, selecione o **Bootstrap method** e clique em **OK**.
